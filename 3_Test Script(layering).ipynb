```python
# 필요한 모듈 import하기
from keras.utils import np_utils
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Activation
import numpy as np
from numpy import argmax
from keras.models import load_model

import tensorflow_hub as hub
```

```python
# 이전에 학습된 모델 불러오기 
model = load_model('emotion_model.h5')
model.summary()
```

```python
import os
from PIL import Image
import glob
```

```python
path = 'datasets_test'
img_li = glob.glob(path+"\\*")

for i in range(len(img_li)):
    img_path = img_li[i]
    cur_img = Image.open(img_path) # 이미지 불러오기
    
    cur_img = cur_img.resize((48,48)) # 사이즈 재정의
    image = np.asarray(cur_img) # ndarray 형태로 변환
    
    try:
        image = np.mean(image,axis = 2)
    except:
        pass
    
    image = image.astype(np.float32)/255 # 실수 형태로 형변환
    
    image = np.reshape(image,(1,48,48)) # 형태 재정의
    preds = model.predict(image) #
    emotion_idx = np.argmax(preds,-1)
    
    if emotion_idx == 0:
        emotion = 'angry'
    elif emotion_idx == 1:
        emotion = 'disgust'
    elif emotion_idx == 2:
        emotion = 'fear'
    elif emotion_idx == 3:
        emotion = 'happy'
    elif emotion_idx == 4:
        emotion = 'neutral'
    elif emotion_idx == 5:
        emotion = 'sad'
    elif emotion_idx == 6:
        emotion = 'surprise'
        
    print('model이 예측한 {}번쨰 사진의 감정은 {}입니다.'.format((i+1),emotion))
```
